# iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
subhajit = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
honey = requests.get(subhajit)

bee = BeautifulSoup(honey.content)

Playersname = [] 
Teamname = []  
Ratingsss = []  

for i in bee.find_all("div",class_='rankings-block__banner--name-large'):
    Playersname.append(i.text)
for i in bee.find_all("div",class_='rankings-block__banner--nationality'):
    Teamname.append(i.text.replace("\n",""))
for i in bee.find_all("div",class_='rankings-block__banner--rating'): 
    Ratingsss.append(i.text)
for i in bee.find_all("td",class_='table-body__cell rankings-table__name name'):
    for j in i.find_all('a'):
        Playersname.append(j.text)
for i in bee.find_all("span",class_='table-body__logo-text'): 
    Teamname.append(i.text)
for i in bee.find_all("td",class_='table-body__cell rating'): 
    Ratingsss.append(i.text)
br=pd.DataFrame({})
br['Player']=Playersname[:10]
br['Team']=Teamname[:10]
br['Rating']=Ratingsss[:10]
br
```

# 6. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
ii) Top 10 women’s ODI players along with the records of their team and rating.
iii) Top 10 women’s ODI all-rounder along with the records of their team and rating.

# i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating

```python
subhajit = "https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
age = requests.get(subhajit)
pro = BeautifulSoup(age.content)
wt = pro.find_all("span",class_='u-hide-phablet')
wtn = []
for i in wt:
    wtn.append(i.text)
wm = [] 
wp = [] 
wr = [] 
wnl = [] 
for i in  pro.find_all("td",class_='rankings-block__banner--matches'): 
    wm.append(i.text)
for i in  pro.find_all("td",class_='rankings-block__banner--points'):
    wp.append(i.text)
for i in  pro.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    wr.append(i.text.replace("\n",""))
for i in  pro.find_all("td",class_='table-body__cell u-center-text'):
    wnl.append(i.text)
for i in range(0,len(wnl)-1,2):
    wm.append(wnl[i])
    wp.append(wnl[i+1])
for i in  pro.find_all("td",class_='table-body__cell u-text-right rating'):
    wr.append(i.text)

wcc=pd.DataFrame({})
wcc['Team_name']=wtn[:10]
wcc['Matches']=wm[:10]
wcc['Points']=wp[:10]
wcc['Ratings']=wr[:10]
wcc
```

# ii)Top 10 women’s ODI players along with the records of their team and rating.

# 7.Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

# 8.Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
subhajit = "https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YQD3CI4zbIU"
weather = requests.get(subhajit)
sanf = BeautifulSoup(weather.content)
po = [] 
sc = []
tr = []
dn = []
for i in sanf.find_all("div", class_="col-sm-2 forecast-label"):
    po.append(i.text)
for i in sanf.find_all('div',class_="col-sm-10 forecast-text"):        
        sc.append(i.text.split(',')[0])
for i in sanf.find_all('div',class_="col-sm-10 forecast-text"):        
        dn.append(i.text)
for i in sanf.find_all('p',attrs={'short-desc'}):
        if i.next_sibling is not None:
            tr.append(i.next_sibling.text)
        else:
            tr.append(' ')
for i in sanf.find_all("div", class_="col-sm-10 forecast-text"):
        dn.append(i.text)
sanfrancisco=pd.DataFrame({})
sanfrancisco['Period']=po[:9]
sanfrancisco['Temperature']=tr[:9]
sanfrancisco['Short_description']=sc[:9]
sanfrancisco['Description']=dn[:9]
sanfrancisco
```

# 9.Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
subhajit = "https://internshala.com/fresher-jobs"
freshers = requests.get(subhajit)
jobs = BeautifulSoup(freshers.content)
jobtitles = []
companynames = []
CTC = []
date = []
titles=jobs.find_all('div',class_="heading_4_5 profile") 
for i in titles:
    jobtitles.append(i.text.replace('\n', ''))   
companies=jobs.find_all('a',class_="link_display_like_text")
for i in companies:
    companynames.append(i.text.strip())
jo = []
info = jobs.find_all("div",class_="item_body")
for i in info:
    jo.append(i.text.replace("\n","").replace("\xa0",""))  
for i in range(2,len(jo),3):
    date.append(jo[i].strip())   
for i in range(1,len(jo),3):
    CTC.append(jo[i].strip())
JOB=pd.DataFrame({})
JOB['Job_Titles']=jobtitles
JOB['Company_Names']=companynames
JOB['CTC']=CTC
JOB['Apply_Date']=date
JOB
```

# 10.Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should include house title, location, area, emi and price

```python
subhajit = "https://www.nobroker.in/property/sale/mumbai/Bandra%20West?searchParam=W3sibGF0IjoxOS4wNTk1NTk2LCJsb24iOjcyLjgyOTUyODcsInBsYWNlSWQiOiJDaElKZTlMNEktSEk1enNSZUdvam1yU1dlVU0iLCJwbGFjZU5hbWUiOiJCYW5kcmEgV2VzdCJ9XQ==&radius=2.0"
nobroker = requests.get(subhajit)
mumbai = BeautifulSoup(nobroker.content)

housetitle = []
location = []
Area = []
EMI = []
price = [] 
houses = mumbai.find_all("h2", class_="heading-6 font-semi-bold nb__1AShY")
for i in houses:
    housetitle.append(i.text)     
loc = mumbai.find_all('div',class_="nb__2CMjv")
for i in loc:
    location.append(i.text)     
area = mumbai.find_all('div',class_="nb__3oNyC")
for i in area:
    Area.append(i.text)
    plot = []
pk = mumbai.find_all('div',class_="font-semi-bold heading-6")
for i in pk:
    plot.append(i.text)
for i in range(1,len(plot),3):
    EMI.append(plot[i])
for i in range(2,len(plot),3):
    price.append(plot[i])
nobroke=pd.DataFrame({})
nobroke['House Title']=housetitle
nobroke['Area']=Area
nobroke['location']=location
nobroke['EMI']=EMI
nobroke['price']=price
nobroke
```
