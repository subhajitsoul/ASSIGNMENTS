# Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %

```python

import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python

driver.get('https://www.flipkart.com')
```

```python
search=driver.find_element_by_class_name("_3704LK")
search.send_keys('Sunglasses')
```

```python
button=driver.find_element_by_class_name("_34RNph")
button.click()
```

```python
brand=[]
product_description=[]
price=[]
discount=[]
```

```python
start=0
end=3
for page in range(start,end):
    bd=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in bd[:100]:
        brand.append(i.text)
    pro_des=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for j in pro_des[:100]:
        product_description.append(j.text)
    prices=driver.find_elements_by_class_name("_30jeq3")
    for k in prices[:100]:
        price.append(k.text) 
    discounts=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    for l in discounts[:100]:
        discount.append(l.text)
    
    nxt_button=driver.find_elements_by_xpath("//a[@class='ge-49M']")
    driver.get(nxt_button[page].get_attribute('href'))
```

```python
sunglass=pd.DataFrame({})
sunglass['Brand']=brand[:100]
sunglass['Product Description']=product_description[:100]
sunglass['Price']=price[:100]
sunglass['Discount']=discount[:100]
sunglass
```

# Q8: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 4 attributes of each sneaker :
1. Brand
2. Product Description
3. Price
4. discount %

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.flipkart.com')
```

```python
cross_btn = driver.find_element_by_xpath("//div[@class = '_2QfC02']/button")
cross_btn.click()
```

```python
search = driver.find_element_by_xpath("//div[@class = '_3OO5Xc']/input")
search.clear()
search.send_keys('sneakers')
search_btn = driver.find_element_by_xpath("//button[@class = 'L0Z3Pu']")
search_btn.click()
time.sleep(3)
```

```python
url_list = []
for j in range(0,3):
    tile = driver.find_elements_by_xpath("//a[@class= '_2UzuFa']")
    for i in tile:
        url_list.append(i.get_attribute('href'))
        next_btn = driver.find_element_by_xpath("//a[@class = '_1LKTO3']/span")
        if next_btn.text == 'Next':
             next_btn.click()
```

```python
brand_name = []
product_description = []
price = []
discount = []
for i in url_list[0:100]:
    driver.get(i)
    time.sleep(1)
    brand = driver.find_element_by_xpath("//span[@class='G6XhRU']")
    brand_name.append(brand.text)
    des = driver.find_element_by_xpath("//span[@class='B_NuCI']")
    product_description.append(des.text)
    pri = driver.find_element_by_xpath("//div[@class = '_30jeq3 _16Jk6d']")
    price.append(pri.text)
    dis = driver.find_element_by_xpath("//div[@class = '_3Ay6Sb _31Dcoz pZkvcx']")
    discount.append(dis.text)
    
driver.close()
```

```python
df = pd.DataFrame({})
df['brand']= brand_name[0:100]
df['description']= product_description[0:100]
df['price']= price[0:100]
df['discount']= discount[0:100]
```

# Q9: Go to the link - https://www.myntra.com/shoes
Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”, as shown in the below image And then scrape First 100 shoes data you get. The data should include “Brand” of the shoes , Short Shoe description, price of the shoe as shown in the below image.

```python
import pandas as pd
import selenium 
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python

driver = webdriver.Chrome("chromedriver.exe")
url='https://www.myntra.com/shoes'
driver.get(url)
```

```python
for i in url:
#for setting price filter 
    pr_ftr=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
    pr_ftr.click()
```

```python
#for setting colour filter 
clr_ftr=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
clr_ftr.click()
```

```python

#for getting the urls related to the next pages
m_urls=[]
url_tags=driver.find_elements_by_xpath('//div[@class="results-showMoreContainer"]//a')
for i in url_tags:
    m_urls.append(i.get_attribute('href'))
m_urls
```

```python
#for getting brands of shoes present across multiple pages
shoe_brands=[]
for i in m_urls:
    driver.get(i)
    b_tag=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
    for i in b_tag:
        shoe_brands.append(i.text)
shoe_brands
```

```python
shoe_price=[]
for i in m_urls:
    driver.get(i)
    price_tags=driver.find_elements_by_xpath('//div[@class="product-price"]')
    for i in price_tags:
        shoe_price.append(i.text)
shoe_price
```

```python

shoe_description=[]
for i in m_urls:
    driver.get(i)
    desc_tags=driver.find_elements_by_xpath('//div[@class="search-searchProductsContainer row-base"]//a')
    for i in desc_tags:
        shoe_description.append(i.get_attribute('href'))
shoe_description
```

```python
sh_desc=[]
for i in shoe_description:
    driver.get(i)        
    d_tag=driver.find_elements_by_xpath('//h1[@class="pdp-name"]')
    for i in d_tag:
        sh_desc.append(i.text)
            
sh_desc
```

```python
m_shoes=pd.DataFrame()
m_shoes['Brand Name']=shoe_brands[0:100]
m_shoes['Shoe Price']=shoe_price[0:100]
m_shoes['Shoe Description']=sh_desc[0:100]
m_shoes
```
