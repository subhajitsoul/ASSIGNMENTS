- 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python

#loading the drivers and URL

driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.maximize_window()
time.sleep(2)
```

```python
#creating the empty list
Rank=[]
Name=[]
Artist=[]
Upload=[]
Views=[]
```

```python
#scraping the Rank
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]"):
    Rank.append(i.text)
```

```python
#scraping the Name
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]"):
    Name.append(i.text.split('[')[0])
```

```python
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]"):
    Artist.append(i.text)
```

```python
#scraping the upload date
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]"):
    Upload.append(i.text)
```

```python
#scraping the number of views
for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]"):
    Views.append(i.text)
```

```python
Yotube_Most_Views={'Rank':Rank,'Name':Name,'Artist':Artist,'Uploaded Date':Upload,'Views':Views}
Yotube_Most_Views=pd.DataFrame(data=Yotube_Most_Views)
Yotube_Most_Views
```

2.- Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#loading the drivers and url
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.bcci.tv/')
driver.maximize_window()

#clickin on International tab
driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']").click()

#clicking on Fixtures
driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").click()

#creating the empty list
Title=[]
Series=[]
Place=[]
Date=[]
Time=[]

#scraping the title
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/strong"):
    Title.append(i.text.replace('\n',' '))

#scraping the series
for i in driver.find_elements_by_xpath("//div[@class='fixture__teams']"):
    Series.append(i.text.replace('\n',' '))

#scraping the place
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    Place.append(i.text.replace('\n',' '))
    
#scraping the date
date= driver.find_elements_by_xpath("//div[@class='fixture__full-date']")
for i in range(len(date)):
    Date.append(date[i].text.split('\n')[0] + date[i].text.split('\n')[1])

#scraping the time
for i in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    Time.append(i.text)
    
#creating the DataFrame
Fixtures={'Title':Title,'Series':Series,'Place':Place,'Date':Date,'Time':Time}
Fixtures=pd.DataFrame(data=Fixtures)
Fixtures
```
